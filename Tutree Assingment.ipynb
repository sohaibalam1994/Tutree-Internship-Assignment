## Identifying and removing the Duplicate values

```python
# Importing all the important libraries
import pandas as pd
import numpy as np
```

```python
# Importing the dataset in csv format using pandas.
data = pd.read_csv("heart-disease.csv")
data.head()
```

```python
# Getting the shape of the dataframe
data.shape
```

```python
# Checking for the duplicate values
data[data.duplicated(keep=False)]
```

```python
# Totalling the all duplicate values.
duplicate = data.duplicated()
sum(duplicate)
```

```python
# Removing the duplicate value by using the function drop_duplicates()
remove_duplicate = data.drop_duplicates()
```

```python
# Now comparing the shape before and after removing the duplicate values.
data.shape, remove_duplicate.shape
```
